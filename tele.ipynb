```python
tel['TotalCharges'] = pd.to_numeric(tel['TotalCharges'], errors='coerce')
```

```python
categorical_features = tel.select_dtypes(include='object').columns

for col in categorical_features:
    plt.figure(figsize=(6,4))
    sns.countplot(data=tel, x=col, hue='Churn')
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
for col in categorical_features:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=tel, x=col, order=tel[col].value_counts().index, palette='Set2')
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
numerical_features = tel.select_dtypes(include=['int64', 'float64']).columns

tel[numerical_features].hist(bins=30, figsize=(10, 10), layout=(len(numerical_features)//2 + 1, 2))
plt.tight_layout()
plt.show()
```

```python
for col in categorical_features:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=tel, x=col, hue='Churn', palette='Set1')
    plt.title(f"{col} vs Churn")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
for col in numerical_features:
    plt.figure(figsize=(6,4))
    sns.boxplot(data=tel, x='Churn', y=col,)
    plt.title(f"Boxplot of {col} by Churn")
    plt.tight_layout()
    plt.show()
```

```python
plt.figure(figsize=(10, 8))
# Select only numerical columns for correlation heatmap
numerical_tel = tel.select_dtypes(include=np.number)
sns.heatmap(numerical_tel.corr(), annot=True,fmt=".2f", cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Create AvgMonthlyCharges feature
tel['AvgMonthlyCharges'] = tel['TotalCharges'] / tel['tenure']
tel['AvgMonthlyCharges'].replace([np.inf, -np.inf], 0, inplace=True)
```

```python
tel['tenure_group'] = pd.cut(tel['tenure'], bins=[0, 12, 24, 48, 60, 72],
                            labels=["0–12", "13–24", "25–48", "49–60", "61–72"])
```

```python
plt.figure(figsize=(4, 4))
tel['Churn'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['skyblue', 'salmon'])
plt.title("Churn Distribution")
plt.ylabel("")
plt.show()
```

```python
df=pd.DataFrame(tel)
Q1=df["tenure"].quantile(0.25)
Q3=df["tenure"].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
tel_clean=tel[(tel["tenure"]>=lower_bound)&(tel["tenure"]<=upper_bound)]
print(tel_clean)
```
